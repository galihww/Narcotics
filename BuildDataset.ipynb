```python
import pandas as pd
```

```python
# Define the file path
file_path = 'data/052.xlsx'

# Read the .xls file into a pandas DataFrame
try:
    df = pd.read_excel(file_path)
    print("File read successfully!")
    # Display the first few rows of the DataFrame
except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
df.head()
```

```python
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
def count_tokens(text: str) -> int:
    return len(tokenizer.encode(text))
```

```python
# Ensure the column exists
if 'Amar Putusan' in df.columns:
    # Fill missing values
    df['Amar Putusan'] = df['Amar Putusan'].fillna('').astype(str)

    # Apply the count_tokens function row-wise
    df['tokens'] = df['Amar Putusan'].apply(count_tokens)
else:
    print("Error: Column 'Amar Putusan' not found in DataFrame")
```

```python
df.rename(columns={"No Putusan": "no_putusan", "Lembaga Peradilan": "lembaga_peradilan", "Barang Bukti": "barang_bukti","Amar Putusan": "amar_putusan"}, inplace=True)
```

```python
def clean_column(df, column_name):
    """
    Cleans the specified column in a DataFrame by removing unnecessary whitespace,
    including double spaces, newlines, leading, and trailing spaces.

    Parameters:
        df (pd.DataFrame): The DataFrame containing the column to clean.
        column_name (str): The name of the column to clean.

    Returns:
        pd.DataFrame: The DataFrame with the cleaned column.
    """
    df[column_name] = df[column_name].astype(str)  # Ensure the column is of string type
    df[column_name] = (
        df[column_name]
        .str.replace('\n', ' ', regex=True)  # Replace newlines with spaces
        .str.replace('\r', '', regex=True)  # Remove carriage returns
        .str.replace('\s+', ' ', regex=True)  # Replace multiple spaces with a single space
        .str.strip()  # Remove leading and trailing spaces
    )
    return df
```

```python
df = clean_column(df, 'barang_bukti')
df = clean_column(df, 'amar_putusan')
```
